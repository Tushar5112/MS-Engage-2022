```python
#Modules Imported
import pandas as pd
import numpy as np
import ast
import pickle
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Loading CSV Files
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
#merging both movies and credits data frame into a single
#data frame i.e. movies keeping joining column to be title
movies = movies.merge(credits,on='title')
```

```python
# filtering unwanted data columns
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
# now checking for any data redundancy
movies.head()
```

```python
# final data frame information
movies.info()
```

```python
# ast module in python converts a list of dictionaries into a single list
import ast
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
# checking for null values
movies.isnull().sum()
```

```python
# dropping null values
movies.dropna(inplace=True)
```

```python
# applying convert function on movies data frame in genres column
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
# same on the keywords column as well
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
# function to filter only top three cast and crew
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
# applying on the cast column
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
# slicing so that it contains only three entries
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
# fetching the director of a movie out the crew
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
# apply on crew column
movies['crew'] = movies['crew'].apply(fetch_director)
movies.sample(5)
```

```python
# combining all the entries
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
# apply on the below columns
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
# sample filtered data
movies.sample(5)
```

```python
# making a new column by combining overview , genres , keywords , cast , crew
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
# sklearn converts a collection of text documents to a matrix of token counts
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
# converting each movie to vector format
vector = cv.fit_transform(new['tags']).toarray()
vector.shape
```

```python
# generates a list of 5000 values
cv.get_feature_names()
# removing redundant names
```

```python
# taking cosine distance between two vectors i.e. movies
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
# renaming the title so that it can be presented as output
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
# recommend fuction to generate movies
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
# sample checking
recommend('Gandhi')
```
